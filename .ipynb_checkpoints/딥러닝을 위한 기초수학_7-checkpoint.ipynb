```python
# Sarsa  
import numpy as np

#  
S = np.array([0, 1, 2, 3])

#        0 1 
A = np.array([0, 1])

#      
R = np.array([[1, -20], [4, -1], [0, 25], [0, 0]])

#   
S1 = np.array([[1, 2], [3, 0], [0, 3], [None, None]])

p = 0.5  #  
alpha = 0.01   # 
gamma = 0.8  # 
n = 3000  #  

#    
Q = np.zeros(R.shape)

#      
def pi(p):
    if np.random.uniform(0, 1) <= p:
        return 0
    else:
        return 1
    
def sarsa():
    s = S[0]
    a = pi(p)
    while S1[s, a] != None:
        a_next = pi(p)
        td = R[s, a] + gamma * Q[S1[s, a], a_next] - Q[s, a]
        Q[s, a] += alpha * td
        s = S1[s, a]
        a = a_next
    print(Q[0, 0], Q[0, 1])
    
#    Sarsa  
for i in range(n):
    sarsa()
```

```python
# Q 
import numpy as np

#  
S = np.array([0, 1, 2, 3])

#        0 1 
A = np.array([0, 1])

#      
R = np.array([[1, -20], [4, -1], [0, 25], [0, 0]])

#   
S1 = np.array([[1, 2], [3, 0], [0, 3], [None, None]])

p = 0.5  #  
alpha = 0.01   # 
gamma = 0.8  # 
n = 3000  #  

#    
Q = np.zeros(R.shape)

#      
def pi(p):
    if np.random.uniform(0, 1) <= p:
        return 0
    else:
        return 1
    
def q_learning():
    s = S[0]
    a = 0
    while S1[s, a] != None:
        a = pi(p)
        max_q = max(Q[S1[s, a], 0], Q[S1[s, a], 1])
        td = R[s, a] + gamma * max_q - Q[s, a]
        Q[s, a] += alpha * td
        s = S1[s, a]
    print(Q[0, 0], Q[0, 1])
    
#    Sarsa  
for i in range(n):
    q_learning()
```
